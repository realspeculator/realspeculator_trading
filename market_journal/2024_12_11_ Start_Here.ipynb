# Portfolio tracking

##### Tools

We are going to use this repo to start tracking our portfolio. We are using Alpaca trading platform as our broker. Alpaca offers great API interface and SDK which we will use to programmatically download our portfolio balances, positions and order history.

We will then upload the positons and orders to Github which will enable us to provide maximum transparency and commit history. Using Github commit history anybody can look back at our trading orders, positions and performance and can feel confident that these trades were done ex-ante and the reporting documents were not changed ex-post.

##### Portfolio Management

Initial portfolio value will be $1million and going forward this value can be tracked [here](https://github.com/realspeculator/realspeculator_trading/tree/master/account_balance_history). Notice the `equity` and `base_value` is currently at 1,000,000 since we have not initiated any trades so far.

The portfolio positions can be tracked [here](https://github.com/realspeculator/realspeculator_trading/tree/master/positions). Please read the disclaimer and use a Paper money account if you wish to replicate the strategies. **DON'T trade with your real money!**

Historical orders can be tracked [here](https://github.com/realspeculator/realspeculator_trading/blob/master/orders/orders_data/2024_12_11_13_40_54_order_history.csv). Using the historical orders anybody can calculate the Profit and Loss generated by this portfolio.

##### Rules

- Max loss on each trade will be 0.005% of the portfolio value. This max loss value will not be exceed under any circumstances
- Target gain on each trade will be 0.0125%. We can move this value around based on market conditions
- We will trade all asset class
- Trades can be long or short
- When a trade is entered we will simultaniously enter the target and stoploss (bracket or OCO order)
- We will try 1 to 2 trades a day depending on the trade setup. But if there is no setup then we will wait patiently
